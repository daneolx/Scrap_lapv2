```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Lista para almacenar todos los DataFrames de cada página
all_vuelos_info = []

# Navegación por paginación
while True:
    # Seleccionar los elementos de vuelos usando selectores CSS
    vuelos = driver.find_elements(By.CLASS_NAME, 'flightDetailsMain')

    # Lista para almacenar la información de los vuelos de la página actual
    vuelos_info = []

    # Iterar sobre los elementos de vuelos encontrados
    for vuelo in vuelos:
        vuelo_data = vuelo.text.strip().split('\n')
        airline_name = None

        # Intentar obtener la aerolínea desde la etiqueta de imagen
        try:
            airline_element = vuelo.find_element(By.CSS_SELECTOR, '.col-airline img')
            src = airline_element.get_attribute('src')
            airline_name = src.split('/')[-1].split('.')[0]
        except:
            pass

        vuelo_data.append(airline_name)
        vuelos_info.append(vuelo_data)

    # Crear un DataFrame para la página actual y añadirlo a la lista
    df_vuelos = pd.DataFrame(vuelos_info)

    # Asignar nombres de columnas
    num_columnas = df_vuelos.shape[1]
    if num_columnas == 7:
        df_vuelos.columns = ['Estado', 'Fecha del Vuelo', 'Hora de Salida', 'Nueva Hora de Salida', 'Destino', 'Vuelo', 'Aerolinea']
    elif num_columnas == 8:
        df_vuelos.columns = ['Estado', 'Fecha del Vuelo', 'Hora de Salida', 'Nueva Hora de Salida', 'Destino', 'Vuelo', 'Gate', 'Aerolinea']
    elif num_columnas == 9:
        df_vuelos.columns = ['Estado', 'Fecha del Vuelo', 'Hora de Salida', 'Nueva Hora de Salida', 'Destino', 'Vuelo', 'Gate', 'Aerolinea', 'ColumnaExtra']
    else:
        raise ValueError(f"Unexpected number of columns: {num_columnas}")

    # Ignorar el primer registro si es necesario
    df_vuelos = df_vuelos.iloc[1:]

    # Agregar el DataFrame de la página actual a la lista
    all_vuelos_info.append(df_vuelos)

    # Navegar a la siguiente página
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, 'nav .pagination .page-item a[aria-label="Next"]')
        next_button.click()
        time.sleep(2)  # Esperar a que la nueva página cargue
    except Exception as e:
        print("No hay más páginas o se produjo un error:", e)
        break

# Concatenar todos los DataFrames en uno solo
final_df_vuelos = pd.concat(all_vuelos_info, ignore_index=True)

# Eliminar registros vacíos o con None
final_df_vuelos.dropna(how='all', inplace=True)

# Mostrar el DataFrame resultante
print(final_df_vuelos)
```
